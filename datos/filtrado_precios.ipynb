```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
industries = pd.read_csv('./datos/industries.csv', sep=';', dtype='str')
companies = pd.read_csv('./datos/us-companies.csv', sep=';', dtype='str')
precios = pd.read_csv('us-shareprices-daily.csv', sep=';')
```

```python
energy_list = industries[industries.Sector=='Energy'].IndustryId.unique()
```

```python
energy_companies = companies[companies.IndustryId.isin(energy_list)]
```

```python
energy_companies
```

```python
energy_prices = precios[precios.Ticker.isin(energy_companies.Ticker)]
```

```python
energy_prices.reset_index(drop=True, inplace=True)
```

```python
energy_prices
```

```python
energy_prices.to_csv('energy_prices.csv')
```
