```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
industries = pd.read_csv('industries.csv', sep=';', dtype='str')
companies = pd.read_csv('us-companies.csv', sep=';', dtype='str')
precios = pd.read_csv('us-shareprices-daily.csv', sep=';')
```

```python
companies = companies.merge(industries, 
                              left_on=('IndustryId'),
                              right_on=('IndustryId'), 
                              how='left')
companies.groupby('Sector', sort=False).count().Ticker.sort_values(ascending=False)
```

```python
energy_companies = companies[companies.Sector=='Energy']
tec_companies = companies[companies.Sector=='Technology']
health_companies = companies[companies.Sector=='Healthcare']
ind_companies = companies[companies.Sector=='Industrials']
```

```python
prices = precios[precios.Ticker.isin(energy_companies.Ticker)]
prices = prices.append(precios[precios.Ticker.isin(tec_companies.Ticker)])
prices = prices.append(precios[precios.Ticker.isin(health_companies.Ticker)])
prices = prices.append(precios[precios.Ticker.isin(ind_companies.Ticker)])
```

```python
prices.reset_index(drop=True, inplace=True)
prices
```

```python
prices.drop(['Open','Low','High','Close','Dividend'], axis=1, inplace=True)
```

```python
prices.to_csv('filtered_prices.csv', compression='bz2', index=False)
```
