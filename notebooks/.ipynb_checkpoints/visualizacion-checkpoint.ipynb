# Visualización

```python
import altair as alt
import pandas as pd
```

```python
resultado = pd.read_csv('../tablas/result_carteras.csv')
```

```python
indice = pd.DataFrame({'Periodo':resultado.Periodo,
                       'Rentabilidad':resultado.Rent_indice,
                       'Sector':resultado.Sector,
                       'Tipo':'Índice'})
resultado.drop('Rent_indice', inplace=True, axis=1)
resultado = resultado.append(indice)
resultado = resultado.drop_duplicates()
resultado
```

```python
tot_cartera=resultado[resultado.Tipo=='Cartera'].groupby('Sector').sum()
tot_indice=resultado[resultado.Tipo=='Índice'].groupby('Sector').sum()
tot_indice['Tipo']='Índice'
tot_cartera['Tipo']='Cartera'
agrupado = tot_cartera.append(tot_indice)
agrupado.reset_index(inplace=True)
```

```python
agrupado
```

```python
input_dropdown_2 = alt.binding_select(options=resultado.Periodo.unique())
select_periodo = alt.selection_single(fields=['Periodo'], bind=input_dropdown_2, name='Resultado del')

input_dropdown_1 = alt.binding_select(options=list(agrupado.Sector.unique()))
select_sector = alt.selection_single(fields=['Sector'], bind=input_dropdown_1, name='Resultado agrupado del')


res_por_periodo=alt.Chart(resultado).mark_bar(size=25).encode(
    x='Tipo',
    y='Rentabilidad',
    color='Tipo',
    column='Sector'
).add_selection(
    select_periodo
).transform_filter(
    select_periodo,
).properties(
    width=130,
    height=250
)

res_agrupado=alt.Chart(agrupado).mark_bar(size=50).encode(
    x='Tipo',
    y='Rentabilidad',
    color='Tipo'
).add_selection(
    select_sector
).transform_filter(
    select_sector,
).properties(
    width=155,
    height=250
)

dash = (res_por_periodo|res_agrupado)
dash
```

```python
dash.save('Rentabilidad.html')
```
