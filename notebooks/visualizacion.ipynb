# Visualización

```python
import altair as alt
import pandas as pd
```

```python
resultado = pd.read_csv('../tablas/result_carteras.csv')
```

```python
indice = pd.DataFrame({'Periodo':resultado.Periodo,
                       'Rentabilidad':resultado.Rent_indice,
                       'Sector':resultado.Sector,
                       'Agrupación':'Índice'})
resultado['Agrupación']='Cartera'
resultado.drop('Rent_indice', inplace=True, axis=1)
resultado = resultado.append(indice)
resultado
```

```python
tot_cartera=resultado[resultado.Agrupación=='Cartera'].groupby('Sector').sum()
tot_indice=resultado[resultado.Agrupación=='Índice'].groupby('Sector').sum()
tot_indice['Agrupación']='Índice'
tot_cartera['Agrupación']='Cartera'
agrupado = tot_cartera.append(tot_indice)
agrupado.reset_index(inplace=True)
```

```python
agrupado
```

```python
list(agrupado.index.unique())
```

```python
input_dropdown_1 = alt.binding_select(options=list(agrupado.Sector.unique()))
select_sector = alt.selection_single(fields=['Sector'], bind=input_dropdown_1, name='Resultado agrupado del')

input_dropdown_2 = alt.binding_select(options=resultado.Periodo.unique())
select_periodo = alt.selection_single(fields=['Periodo'], bind=input_dropdown_2, name='Resultado del')


res_por_periodo=alt.Chart(resultado).mark_bar().encode(
    x='Agrupación',
    y='Rentabilidad',
    color='Agrupación',
    column='Sector'
).add_selection(
    select_periodo
).transform_filter(
    select_periodo,
).properties(
    width=175
)

res_agrupado=alt.Chart(agrupado).mark_bar().encode(
    x='Agrupación',
    y='Rentabilidad',
    color='Agrupación'
).add_selection(
    select_sector
).transform_filter(
    select_sector,
).properties(
    width=175
)

res_por_periodo&res_agrupado
```
